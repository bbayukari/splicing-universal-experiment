```python
import pandas as pd

df = pd.read_csv("/home/wangzz/github/splicing-universal-experiment/JMLR_reponse/ic.csv")
df2 = pd.read_csv("/home/wangzz/github/splicing-universal-experiment/JMLR_reponse/ic_lasso.csv")

df = pd.concat([df, df2])
```

```python
df.groupby(["model", "method"]).count()
```

```python
df.groupby(["model", "method"])["time"].apply(lambda x: f"{x.mean().round(2)}({x.std().round(2)})")
```

```python
# Recall
df.groupby(["model", "method"])["accuracy"].apply(lambda x: f"{x.mean().round(2)}({x.std().round(2)})")
```

```python
# Precision
df.groupby(["model", "method"])[["accuracy", "support_size", "sparsity_level"]].apply(
    lambda x: f'{(x["accuracy"]*x["sparsity_level"] / x["support_size"]).mean().round(2)}({(x["accuracy"]*x["sparsity_level"] / x["support_size"]).std().round(2)})'
    )
```
