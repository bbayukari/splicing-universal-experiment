```python
import pandas as pd

data = []
with open('ic_supp.txt', 'r') as f:
    for l in f.readlines():
        if not l.startswith('{'):
            continue
        data.append(eval(l))
df = pd.DataFrame(data)  
```

```python
df.groupby(['method', 'model'])['time'].apply(lambda x: f"{x.mean().round(2)}({x.std().round(2)})").unstack()
```

```python
df.groupby(['method', 'model'])['accuracy'].apply(lambda x: f"{x.mean().round(2)}({x.std().round(2)})").unstack()
```

```python
# Precision
df.groupby(['method', 'model'])[["accuracy", "support_size", "sparsity_level"]].apply(
    lambda x: f'{(x["accuracy"]*x["sparsity_level"] / x["support_size"]).mean().round(2)}({(x["accuracy"]*x["sparsity_level"] / x["support_size"]).std().round(2)})'
    ).unstack()
```
