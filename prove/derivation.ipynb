```python
from sympy import *

s, m = symbols('s m')

delta = 1/4/m * (1 - 4/(s-2)*m*m)
epsilon = 1/(1/2/delta/m -1)
l = (1-sqrt(1-4/(s-2)*((1+epsilon)*m)**2)) / (4/(s-2)*((1+epsilon)*m)**2)
```

```python
eq1 = epsilon/m/(1+epsilon)**2 - delta*(2-4*m*delta)
eq2 = 2*l*(1+epsilon)**2/epsilon*m*delta
```

```python
simplify(eq1)
```

```python
simplify(eq2)
```

```python
x, y = symbols('x y') # x = m**2, y = s-2
eq2_simple = -1/8/x*(sqrt(1-x*y/(x+y/4)**2)-1)*(4*x+y)
simplify(eq2_simple)
```
