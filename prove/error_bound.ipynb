abess-universal 参数估计的L2误差的界

```python
import math
import matplotlib.pyplot as plt  
import numpy as np
```

```python
def f(x):
    return (x+15.52)/14/math.sqrt(0.5 - 9/98*(x+2)*(1.42-x)**2 - 3/14*x*(x+1)*(1.42-x) - 1/8*(x-1)**2*(x+2))
```

```python
x = np.linspace(1,1.42,1000)
y = np.array([f(i) for i in x])
plt.plot(x,y)
```
